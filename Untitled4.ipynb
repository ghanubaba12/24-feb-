```python
Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.
ans-
Sure! Here's an example function that re-indexes a DataFrame with a new index that starts from 1 and increments by 2 for each row:

python
Copy code
import pandas as pd

def reindex_dataframe(df):
    """
    Re-indexes a DataFrame with a new index that starts from 1 and increments by 2 for each row.
    
    Parameters:
        df (pandas.DataFrame): The DataFrame to re-index.
        
    Returns:
        pandas.DataFrame: The re-indexed DataFrame.
    """
    # Create a new index that starts from 1 and increments by 2 for each row
    new_index = pd.RangeIndex(start=1, stop=2*len(df), step=2)
    
    # Re-index the DataFrame using the new index
    df = df.reindex(new_index)
    
    return df
You can then call this function with your DataFrame as an argument:

python
Copy code
# Create a sample DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# Re-index the DataFrame using the reindex_dataframe function
df = reindex_dataframe(df)

# Print the resulting DataFrame
print(df)
This will output a DataFrame with the same columns as the input DataFrame, but with a new index that starts from 1 and increments by 2 for each row.




```

```python
Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.
ans-
Sure, here's an example function that iterates over a DataFrame and calculates the sum of the first three values in the 'Values' column:

python
Copy code
def sum_first_three_values(df):
    """
    Iterates over a DataFrame and calculates the sum of the first three values in the 'Values' column.
    
    Parameters:
        df (pandas.DataFrame): The DataFrame to iterate over.
        
    Returns:
        None
    """
    # Initialize a counter and a sum variable
    count = 0
    total = 0
    
    # Iterate over the rows in the DataFrame
    for index, row in df.iterrows():
        
        # If the counter is less than 3, add the value in the 'Values' column to the total
        if count < 3:
            total += row['Values']
            count += 1
            
        # If the counter is greater than or equal to 3, break out of the loop
        else:
            break
    
    # Print the total to the console
    print(f"The sum of the first three values in the 'Values' column is: {total}")
You can then call this function with your DataFrame as an argument:

python
Copy code
# Create a sample DataFrame
df = pd.DataFrame({'Values': [1, 2, 3, 4, 5, 6]})

# Call the sum_first_three_values function with the DataFrame
sum_first_three_values(df)
This will output the sum of the first three values in the 'Values' column to the console.




```

```python
Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.
ans-
Sure, here's an example function that creates a new column 'Word_Count' in a DataFrame based on the number of words in the 'Text' column:

python
Copy code
def add_word_count_column(df):
    """
    Creates a new column 'Word_Count' in a DataFrame based on the number of words in the 'Text' column.
    
    Parameters:
        df (pandas.DataFrame): The DataFrame to add the new column to.
        
    Returns:
        pandas.DataFrame: The DataFrame with the new 'Word_Count' column added.
    """
    # Split the 'Text' column into words using whitespace as the delimiter
    words = df['Text'].str.split()

    # Get the number of words in each row and assign it to the 'Word_Count' column
    df['Word_Count'] = words.apply(len)

    # Return the DataFrame with the new column added
    return df
You can then call this function with your DataFrame as an argument:

python
Copy code
# Create a sample DataFrame
df = pd.DataFrame({'Text': ['This is a sample sentence.', 'Here is another sentence.', 'And one more.']})

# Call the add_word_count_column function with the DataFrame
df = add_word_count_column(df)

# Print the updated DataFrame
print(df)
This will output the updated DataFrame with the new 'Word_Count' column added to the console:

vbnet
Copy code
                          Text  Word_Count
0   This is a sample sentence.           5
1    Here is another sentence.           4
2                And one more.           3




```

```python
Q5. How are DataFrame.size() and DataFrame.shape() different?
ans-
Both DataFrame.size() and DataFrame.shape() are methods used to get information about the dimensions of a Pandas DataFrame, but they provide different types of information:

DataFrame.size() returns the total number of elements in the DataFrame, which is equal to the product of the number of rows and columns in the DataFrame.

DataFrame.shape() returns a tuple containing the number of rows and columns in the DataFrame.

Here's an example to illustrate the difference:

python
Copy code
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# Get the size of the DataFrame
size = df.size

# Get the shape of the DataFrame
shape = df.shape

# Print the results
print(f"Size: {size}")  # Output: Size: 9
print(f"Shape: {shape}")  # Output: Shape: (3, 3)
In this example, the df.size method returns the total number of elements in the DataFrame, which is 9. The df.shape method returns a tuple containing the number of rows and columns in the DataFrame, which is (3, 3).





```

```python
Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
ans-
To create a new column in a Pandas DataFrame that contains only the username part of each email address in another column, we can use the apply() method with a lambda function that extracts the username from the email address using the split() method.

Here's an example:

python
Copy code
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'Email': ['jdoe@example.com', 'msmith@example.com', 'jsmith@example.com']})

# Create a new column 'Username' that contains only the username part of each email address
df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])

# Print the DataFrame
print(df)
In this example, we create a new column 'Username' in the DataFrame df that contains only the username part of each email address in the 'Email' column. We use the apply() method to apply a lambda function to each element of the 'Email' column, which splits the email address at the '@' symbol and returns the first part. The result is a DataFrame with two columns: 'Email' and 'Username'.





```
